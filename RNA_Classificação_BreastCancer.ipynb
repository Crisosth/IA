```python
#Importação da biblioteca pandas
import pandas as pd
```

```python
#base de dados =  previsores + classe

#criação atributos previsores - 30 colunas x 569 registros
previsores = pd.read_csv('entradas_breast.csv')

#criação classe - 1 coluna x 569 registros
classe = pd.read_csv('saidas_breast.csv')
```

```python
#importação da função que faz a divisão da base de dados
from sklearn.model_selection import train_test_split
```

```python
#divisão da base de dados - treinamento(75% = 426 registros) - teste (25% = 143 registros)
#criação das variaveis que vão receber as partes da base de dados divididas
#previsores_treinamento = 38 colunas x 426
#previsores_teste = 30 colunas x 143 registros
#classe_treinamento = 1 coluna x 426 registros
#classe_teste = 1 coluna x 143 registros

previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25)
```

```python
previsores_treinamento
```

```python
# importação da biblioteca keras
import keras

# importação do modelo sequencial
# sequencia de camadas: Entrada-Oculta-Saída
from keras.models import Sequential

# importação da classe de camadas densas
from keras.layers import Dense
```

```python
# criando a rede neural do tipo sequencial(objeto)
# nome = classificador
classificador = Sequential()

# adicionando a primeira camada oculta
classificador.add(Dense(units=16, activation= 'relu',
                        kernel_initializer = 'random_uniform', input_dim= 30))
# adicionando a segunda camada oculta
classificador.add(Dense(units = 12, activation= 'relu', kernel_initializer= 'random_uniform'))

# adicionando a terceira camada oculta
classificador.add(Dense(units = 3, activation= 'relu', kernel_initializer= 'random_uniform'))
```

```python
# adicionando a camada de saída
classificador.add(Dense(units=1, activation= 'sigmoid'))
```

```python
#compilando a rede neural com o metodo compile
otimizador = keras.optimizers.Adam(learning_rate= 0.001, weight_decay= 0.0211, clipvalue=25.5)
classificador.compile(optimizer = otimizador, loss= 'binary_crossentropy', metrics = ['binary_accuracy'])
#classificador.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
```

```python
# treinando a Rede Neural
classificador.fit(previsores_treinamento, classe_treinamento, batch_size = 150, epochs = 155)
```

```python
# executando a Rede Neural com a Base de Dados
# previsores_teste = 30 colunas x 143 registros
previsoes = classificador.predict(previsores_teste)
```

```python
previsoes
```

```python
#convertendo as previsoes da Rede Neural para V ou F
previsoes = (previsoes > 0.5)
previsoes
```

```python
# avaliando o acerto - comparar previsoes e classe_teste
#importando o método de comparação
from sklearn.metrics import accuracy_score

#avaliando a precisão das previsoes com a classe teste
precisao = accuracy_score(classe_teste, previsoes)
```

```python
#mostrando a precisao
precisao
```

```python
# Avaliandoi o acerto - comparar previsores e classe_teste
#usando recursos do Keras
resultado= classificador.evaluate(previsores_teste, classe_teste)
resultado
```
