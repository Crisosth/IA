```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
dados = pd.read_csv("Doencas_cardiovasculares.csv", sep = ',')
dados.head()
```

```python
labelencoder = LabelEncoder()
dados = dados.apply(labelencoder.fit_transform)
dados.head()
#feita a normalização dos dados
```

```python
X = dados[[ 'age', 'sex', 'chest pain type', 'resting bp s', 'cholesterol', 'fasting blood sugar', 'max heart rate', 'exercise angina', 'ST slope']]
Y = dados[['target']]
#feita a separação da taxonomia
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, shuffle=True,test_size=0.1, random_state=0 )

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test =  scaler.transform(X_test)
#separação dos dados para treino e teste
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
y_pred = knn.predict(X_test)
```

```python
print(metrics.classification_report(Y_test, y_pred))
```
