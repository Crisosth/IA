```python
# corrigindo valores faltantes
# criando variavel com as associações dos campos/valores
valores = {'vehicleType': 'limousine',
           'gearbox': 'manuell',
           'model': 'golf',
           'fuelType': 'benzin',
           'notRepairedDamage': 'nein'}
```

```python
# fazendo as correções
base = base.fillna(value = valores)
```

```python
#separando a base dos atributos previsores
# iloc [intervalo de linhas (0 :f+1), intervalo de colunas(0: f+1)]
previsores = base.iloc[:, 1:12] .values
```

```python
# transformando valores categoricos em valores numericos
# importando a biblioteca LabelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# criando o objeto da classe Label Encoder
labelencoder_previsores = LabelEncoder()
```

```python
# transformando o atributo abtest (coluna 0) em valores numericos
previsores[:, 0] = labelencoder_previsores.fit_transform(previsores[:, 0])
```

```python
# transformando os outros atributos (coluna 1, 2 , 3 , 4, 5, 8, 9 , 10) em valores numericos
previsores[:, 1] = labelencoder_previsores.fit_transform(previsores[:, 1])
previsores[:, 3] = labelencoder_previsores.fit_transform(previsores[:, 3])
previsores[:, 5] = labelencoder_previsores.fit_transform(previsores[:, 5])
previsores[:, 8] = labelencoder_previsores.fit_transform(previsores[:, 8])
previsores[:, 9] = labelencoder_previsores.fit_transform(previsores[:, 9])
previsores[:, 10] = labelencoder_previsores.fit_transform(previsores[:, 10])
```

```python
previsores[0:20]
```

```python
# detalhando os valores convertidos oelo LabelEncoder
# importando a biblioteca OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# criando o onjeto da classe One Hot Encoder
# detalhando os valores categoricos das colunas convertidas
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10])], remainder='passthrough')
```

```python
# atualizando a base de dados previsores
previsores = columnTransformer.fit_transform(previsores).toarray()
```

```python
# importando bibliotecas
from keras.models import Sequential
from keras.layers import Dense
```

```python
# criando a RNA regressor
regressor = Sequential()
```

```python
# criando a primeira camada oculta e a camada de entrada
# numero de neuronios camada oculta = (316 + 1) / 2 = 158.5
# numero de neuronios camada entrada = 316
regressor.add(Dense(units = 158, activation = 'relu', input_dim = 316))

# criando mais uma camada oculta
regressor.add(Dense(units = 158, activation = 'relu'))

# criando a camada de saida
# função de ativação linear para previsão de números
regressor.add(Dense(units = 1, activation = 'linear'))
```

```python
# compilação da RNA
# funções loss e metrics especificas para regressão - mean_absolute_error
regressor.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['mean_absolute_error'])
```

```python
# treinamento da RNA
# atualização de pesos a cada 300 registros
# repetir 100 vezes
regressor.fit(previsores, preco_real, batch_size = 300, epochs = 100)
```

```python
# fazendo as previsões dos preços dos carros
# vai ler cada um dos 359291 registros da base de dados previsores
# passar na RNA com os últimos pesos calculados
# e vai prever novos preços dos carros
previsoes = regressor.predict(previsores)
```

```python
# verificando as medias dos preços previstos
prev = previsoes.mean()
```

```python
# verificando as medias dos preços existentes
real = preco_real.mean()
```

```python
# avaliando a precisao da RNA - comparar previsoes e preco_real
prec = (prev * 100) / real
```
