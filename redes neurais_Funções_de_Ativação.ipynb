```python
#importação da biblioteca matematica
import numpy as np
```

**Função de Ativação: Sigmoid**
#####Esta função é comumente usada em problemas de classificação binária, como em redes neurais para prever se algo é verdadeiro ou falso.

```python
#Gráfico da função Sigmoid:
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
sigmoid = 1 / (1 + np.exp(-x))

plt.plot(x, sigmoid)
plt.title('Função de Ativação Sigmoid')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.show()
```

```python
#Representação da função Sigmoid em Python:
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

#Chamada para execução da função Sigmoid:
resultado_sigmoid = sigmoid(1.0)

#Impressão do valor retornado pela função Sigmoid:
print("Resultado da função Sigmoid:", resultado_sigmoid)
```

**Função de Ativação: Degrau**

**Função de Ativação: Tangente Hiperbólica**
#####A função tangente hiperbólica é frequentemente usada em problemas de classificação e regressão em redes neurais.


```python
#Gráfico da função Tangente Hiperbólica:
x = np.linspace(-5, 5, 100)
tanh = np.tanh(x)

plt.plot(x, tanh)
plt.title('Função de Ativação Tangente Hiperbólica')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.grid(True)
plt.show()
```

```python
#Representação da função Tangente Hiperbólica em Python:
def tangente_hiperbolica(x):
    return np.tanh(x)

#Chamada para execução da função Tangente Hiperbólica:
resultado_tanh = tangente_hiperbolica(1.0)

#Impressão do valor retornado pela função Tangente Hiperbólica:
print("Resultado da função Tangente Hiperbólica", resultado_tanh)
```

**Função de Ativação: ReLU (Rectified Linear Unit)**
#####A função ReLU é comumente utilizada em redes neurais profundas e é eficaz na maioria dos casos, especialmente em problemas de classificação e regressão.

```python
#Gráfico da função ReLU:
x = np.linspace(-5, 5, 100)
relu = np.maximum(0, x)

plt.plot(x, relu)
plt.title('Função de Ativação ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.show()
```

```python
#Representação da função ReLU em Python:
def relu(x):
    return np.maximum(0, x)

#Chamada para execução da função ReLU:
resultado_relu = relu(2.0)

#Impressão do valor retornado pela função ReLU:
print("Resultado da função ReLU para x = 2.0:", resultado_relu)
```

**Função de Ativação: Linear**
#####A função linear é utilizada principalmente em modelos de regressão simples, onde se espera uma resposta proporcional à entrada.

```python
#Gráfico da função Linear:
x = np.linspace(-5, 5, 100)
linear = x

plt.plot(x, linear)
plt.title('Função de Ativação Linear')
plt.xlabel('x')
plt.ylabel('Linear(x)')
plt.grid(True)
plt.show()
```

```python
#Representação da função Linear em Python:
def linear(x):
    return x

#Chamada para execução da função Linear:
resultado_linear = linear(10)

#Impressão do valor retornado pela função Linear:
print("Resultado da função Linear para x = 10:", resultado_linear)

```

**Função de Ativação: Softmax**
#####A função Softmax é comumente usada na camada de saída de redes neurais para problemas de classificação multiclasse.

```python
#Representação da função Softmax em Python:
def softmax(x):
    exp_values = np.exp(x - np.max(x))
    return exp_values / exp_values.sum(axis=0)

#Chamada para execução da função Softmax:
valores = np.array([2.0, 1.0, 0.1])
resultado_softmax = softmax(valores)

#Impressão do valor retornado pela função Softmax:
print("Resultado da função Softmax para os valores:", resultado_softmax)
```
