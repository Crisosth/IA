```python
# Importando as bibliotecas necess√°rias:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
# Carregando a base de dados:
df = pd.read_csv('doencas_cardiovasculares.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, LabelBinarizer
labelencoder = LabelEncoder()
df = df.apply(labelencoder.fit_transform)
df.head()
#feita a normnaliza√ß√£o dos dados
```

```python
# X = dados de entrada Y = classe
Y = df['target'].values
X = df.drop(columns = ['target'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, shuffle = True, test_size=0.1,random_state=0)
# Verificando as formas dos dados:
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
# Inst√¢nciando o objeto classificador:
clf = DecisionTreeClassifier()
```

```python
# Treinando o modelo de arvore de decis√£o:
clf = clf.fit(X_train,y_train)
```

```python
# Verificando as features mais importantes para o modelo treinado:
clf.feature_importances_
```

```python
for feature,importancia in zip(X.columns,clf.feature_importances_):
    print("{}:{}".format(feature, importancia))
##Lista os campos com mais import√¢ncia para mostrar quais deles tem maior relev√¢ncia
```

```python
resultado = clf.predict(X_test)
resultado
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,resultado))
#exibi√ß√£o das taxas de precis√£o
```
